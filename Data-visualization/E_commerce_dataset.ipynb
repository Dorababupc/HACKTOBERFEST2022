###4. Data Visualization

Order ID seems to be a common column and is present in both order_details and list of orders csv. So we perform a left join and start doing more analysis

```python
df1 = pd.merge(order_detailsDF, orderListDF, on='Order ID', how='left')
```

```python
df1.head()
```

**Total amount spent by each customer:**

```python
Amtspent = df1.groupby(['CustomerName','State'])['Amount'].sum().reset_index()
Amtspent
```

```python
# Sorting top 10 customers
S1 = Amtspent.sort_values(by='Amount', ascending=False).head(10)
S1
```

**Plot on customers who have spent the most amount**

```python
fig = px.bar(S1, x='CustomerName', y='Amount', color='CustomerName', barmode='relative', hover_data=['State'])
fig.show()
```

Yaanvi has spent most of her amount

**Countplot on number of items sold with respect to each state and category**

```python
px.histogram(df1, x='State', color='Category', text_auto=True)
```

Maximum number of orders were sold in Madhya Pradesh and Maharashtra
Least in Tamil Nadu

**Grouping by States with respect to Total number of orders**

```python
Total_order = df1.groupby(['State'])['Category'].count().reset_index()
Total_order
```

**Top 10 states with most number of purchased items/ orders**

```python
# Top 10 states with most number of purchased items/ orders
top_10_states = Total_order.sort_values(by='Category', ascending=False).head(10)
```

```python
fig2 = px.bar(top_10_states, x='State', y='Category', color='State', barmode='relative', hover_data=['State'])
fig2.show()
```

Maharashtra has most number of purchased items and Karnataka has the least

**Adding a new column called Region and differentiating the states accordingly**

```python
#Creating a new column called Region 
North = ["Jammu And Kashmir", "Punjab","Himachal Pradesh", "Delhi","Uttarakhand","Haryana","Uttar Pradesh","Chandigarh"]
South = ["Kerala","Tamil Nadu", "Andhra Pradesh","Karnataka"]
West = ["Maharashtra", "Gujarat","Goa","Rajasthan"]
East = ["Nagaland","Sikkim","West Bengal","Bihar"]
Central = ["Madhya Pradesh"]
```

```python
orderListDF.State.unique()
```

```python
# Function for region
x=""
def region(x):
    if x in East:
        return "East"
    if x in South:
        return "South"
    if x in West:
        return "West"
    if x in North:
        return "North"
    if x in Central:
        return "Central"
df1['Region'] = df1['State'].apply(region)
```

```python
df1.tail()
```

**Pie chart to plot which Region has the most number of Orders**

```python
import plotly.express as px
px.pie(df1, values='Amount', names='Region',color_discrete_sequence=px.colors.sequential.RdBu)
```

According to piechart West has the most number of orders, followed by Central and North India and South has the least  number of orders.
